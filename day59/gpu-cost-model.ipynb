## 1. Cost Model

$/token = GPU_cost / (tokens_per_second * 3600 * utilization).


```python
gpus = [("A100 80GB",3.5,312,2.0,80),("H100 80GB",10.0,989*2,3.35,80),
        ("GB10 Spark",0.5,67*2,0.273,128)]
utilizations = [0.5, 0.7, 0.9]
print(f"{'GPU':<15} {'$/GPU/hr':>9}", end='')
for u in utilizations: print(f" {u:.0%} util $/1Mtok", end='')
print()
for name,cost,tflops,bw,mem in gpus:
    # decode throughput ~ bw limited
    tps = bw*1e12 / (4*4096**2*2) * 1000  # rough tokens/s
    print(f"{name:<15} {cost:>9.1f}", end='')
    for u in utilizations:
        cpt = cost / (tps*u*3600) * 1e6
        print(f" {cpt:>13.3f}", end='')
    print()
```
