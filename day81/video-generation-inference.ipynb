## Concept Overview

Video generation (Sora-class) runs diffusion over 3D latent volumes (time x height x width). Context parallelism splits the temporal dimension across GPUs, with attention spanning the full sequence via ring attention.

## 1. Video Inference Compute

A 10-second 720p video at 24fps = 240 frames. 8x8 spatial compression -> 30x90 latent grid. 4x temporal compression -> 60 time steps. Sequence length: 60*30*90 = 162,000 tokens.


```python
import numpy as np

def video_sequence_length(seconds, fps=24, H=720, W=1280,
                          spatial_compress=8, temporal_compress=4):
    frames = seconds * fps
    lat_frames = int(np.ceil(frames / temporal_compress))
    lat_h = H // spatial_compress
    lat_w = W // spatial_compress
    seq_len = lat_frames * lat_h * lat_w
    return seq_len, lat_frames, lat_h, lat_w

print("Video generation sequence lengths:")
print(f"{'Duration':>10} {'Frames':>8} {'Latent Seq':>12} {'Attn Flops (B)':>16}")
for secs in [2, 5, 10, 30, 60]:
    seq, lf, lh, lw = video_sequence_length(secs)
    attn_flops = 2 * seq**2  # QK^T flops
    print(f"{secs:>10}s {secs*24:>8} {seq:>12,} {attn_flops/1e9:>16,.0f}")
print()
print("Context parallelism splits seq_len across N GPUs.")
print("Ring attention: each GPU holds seq_len/N tokens, passes KV around ring.")
```
