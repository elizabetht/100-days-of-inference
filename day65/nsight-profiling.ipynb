## 1. Nsight Systems Usage

nsys profile captures the complete execution timeline, enabling identification of kernel launch gaps, memory copies, and synchronization stalls.


```python
print("Nsight Systems profiling commands:")
print()
print("Record:")
print("  nsys profile -w true -t cuda,nvtx,osrt,cudnn,cublas \\")
print("    --cuda-memory-usage true \\")
print("    -o profile_output \\")
print("    python inference_server.py")
print()
print("View:")
print("  nsys-ui profile_output.nsys-rep")
print()
print("Key things to look for:")
analysis = [("Kernel launch gaps > 50us", "CPU overhead — reduce kernel count"),
             ("Memory transfer on critical path", "Move transfers off critical path"),
             ("cudaStreamSynchronize calls", "Unnecessary synchronization"),
             ("SM utilization < 50%", "Memory-bound or occupancy issue"),
             ("Small kernel durations < 10us", "Launch overhead dominant — fuse ops")]
for issue, action in analysis:
    print(f"  {issue:<45} → {action}")
```
