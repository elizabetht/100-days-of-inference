## Concept Overview

Whisper processes 30-second audio chunks. For real-time transcription, latency = chunk_duration + model_inference_time. Streaming ASR breaks audio into smaller chunks with acoustic overlap to reduce latency.

## 1. Whisper Latency Model

For streaming, chunk size drives latency vs accuracy tradeoff.


```python
import numpy as np

def whisper_latency(audio_s, chunk_s, model_rtf=0.05):
    """RTF = real-time factor = inference_time/audio_duration."""
    n_chunks = int(np.ceil(audio_s / chunk_s))
    inference_per_chunk = chunk_s * model_rtf
    total_latency = chunk_s + inference_per_chunk  # perceived latency
    total_time = n_chunks * inference_per_chunk
    return total_latency, total_time

print("Whisper streaming: latency vs accuracy tradeoff")
print(f"{'Chunk size':>12} {'Latency (s)':>14} {'RTF factor':>12}")
for chunk in [0.5, 1.0, 2.0, 5.0, 10.0, 30.0]:
    lat, total = whisper_latency(60.0, chunk)
    print(f"{chunk:>12.1f} {lat:>14.2f} {total/60.0:>12.2f}x")
print()
print("Optimizations:")
for opt in ["Distil-Whisper: 6x faster, 1% WER increase", "INT8 quantization: 2x faster", "CTranslate2 backend: 4x faster than transformers"]:
    print(f"  - {opt}")
```
