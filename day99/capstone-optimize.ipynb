## 1. Bottleneck Analysis Workflow

Systematic bottleneck identification: start from end-to-end metrics, drill down to component-level profiling.


```python
import numpy as np

# Simulated benchmark results from deployed stack
benchmark_results = {
    "concurrency": [1, 4, 8, 16, 32, 64],
    "ttft_p99_ms": [120, 145, 220, 580, 1400, 3200],
    "throughput_tps": [45, 170, 310, 480, 520, 515],
    "gpu_util_pct": [12, 45, 78, 95, 97, 98],
    "kv_cache_util_pct": [5, 18, 35, 68, 82, 94],
    "queue_depth": [0, 0, 0, 2, 15, 45],
}

print("Benchmark results vs concurrency:")
print(f"{'Conc':>6} {'TTFT P99':>10} {'Throughput':>12} {'GPU%':>6} {'KV%':>6} {'Queue':>7}")
for i, c in enumerate(benchmark_results["concurrency"]):
    t = benchmark_results["ttft_p99_ms"][i]
    tp = benchmark_results["throughput_tps"][i]
    g = benchmark_results["gpu_util_pct"][i]
    k = benchmark_results["kv_cache_util_pct"][i]
    q = benchmark_results["queue_depth"][i]
    bottleneck = " <- BOTTLENECK" if q > 10 else ""
    print(f"{c:>6} {t:>9}ms {tp:>11} {g:>5}% {k:>5}% {q:>7}{bottleneck}")

print()
print("Diagnosis: KV cache saturation at concurrency=32+ causes queue buildup.")
print("Fix options:")
print("  1. Reduce max_model_len from 8192 to 4096 (frees KV memory)")
print("  2. Enable KV cache quantization (INT8 KV = 2x cache capacity)")
print("  3. Add a second replica (scale horizontally)")
print("  4. Implement prefix caching to improve KV hit rate")
print()
print("Applied fix: KV INT8 quantization + prefix caching")
print("Result: P99 TTFT at c=32 drops from 1400ms to 680ms (+2x improvement)")
```
