## Concept Overview

Knowledge distillation trains a small student model to match a large teacher's output distribution. The inference tradeoff: student runs at 10x lower latency but with degraded quality. Speculative decoding can recover quality by using teacher as verifier.

## 1. Teacher-Student Tradeoff

Model quality degradation as function of student size relative to teacher.


```python
import numpy as np, matplotlib.pyplot as plt

def quality_latency_tradeoff(teacher_params_b=70, student_configs=None):
    """Model quality-latency tradeoff for distillation."""
    if student_configs is None:
        student_configs = [(1, 0.92), (3, 0.95), (7, 0.97), (13, 0.98)]
    # Base: teacher quality=1.0, latency=1.0
    results = [(teacher_params_b, 1.0, 1.0)]  # (params, quality, speedup)
    for params, quality_frac in student_configs:
        speedup = teacher_params_b / params  # rough latency scaling
        results.append((params, quality_frac, speedup))
    return results

results = quality_latency_tradeoff()
print("{:>10} {:>10} {:>10} {:>10}".format("Model","Quality","Speedup","Pareto?"))
for params, quality, speedup in results:
    pareto = "YES" if quality > 0.95 and speedup > 5 else ""
    label = str(params)+"B" if params < 70 else "70B (teacher)"
    print("{:>10} {:>10.2f} {:>9.1f}x {:>10}".format(label,quality,speedup,pareto))

fig, ax = plt.subplots(figsize=(8,5))
params = [r[0] for r in results]
qualities = [r[1] for r in results]
speeds = [r[2] for r in results]
scatter = ax.scatter(speeds, qualities, c=params, s=100, cmap="viridis")
ax.set_xlabel("Speedup vs teacher"); ax.set_ylabel("Relative quality")
ax.set_title("Distillation: quality vs speedup tradeoff")
plt.colorbar(scatter, label="Params (B)"); plt.grid(True)
plt.savefig("distillation_tradeoff.png", dpi=100); plt.show()
```
