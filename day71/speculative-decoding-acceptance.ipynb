## 1. Acceptance Rate Simulation

Acceptance rate depends on draft quality and vocabulary entropy. Common tokens (short, frequent) accept more than rare tokens.


```python
import numpy as np
np.random.seed(42)

def acceptance_rate(draft_quality, vocab_size=32000, n_samples=10000):
    """Simulate acceptance rate as function of draft quality."""
    accepts = 0
    for _ in range(n_samples):
        target_probs = np.random.dirichlet(np.ones(100))
        # Draft is noisy version of target
        noise = np.random.dirichlet(np.ones(100)) * (1-draft_quality)
        draft_probs = target_probs * draft_quality + noise
        draft_probs /= draft_probs.sum()
        tok = np.random.choice(100, p=draft_probs)
        accept_prob = min(1.0, target_probs[tok]/(draft_probs[tok]+1e-8))
        if np.random.random() < accept_prob: accepts += 1
    return accepts / n_samples

print("Acceptance rate vs draft quality:")
for dq in [0.5, 0.6, 0.7, 0.8, 0.9]:
    ar = acceptance_rate(dq)
    print(f"  quality={dq:.1f}: acceptance={ar:.2%}")
print()
print("In practice: Llama-3-8B as draft for Llama-3-70B → ~80% acceptance")
```
