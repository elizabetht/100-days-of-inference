## Concept Overview

Medusa adds K parallel draft heads to a frozen base model. Head k predicts token t+k+1 conditioned on the shared hidden state from position t. Tree attention verifies all K draft tokens simultaneously.

## 1. Medusa Tree Attention

With K=5 heads and gamma draft tokens each, we get a tree of candidate continuations. Tree attention verifies all branches in a single forward pass.


```python
import numpy as np

def medusa_tree_candidates(n_heads=5, top_k=5):
    """Count candidate trees generated by Medusa."""
    return top_k ** n_heads

def medusa_speedup(acceptance_rates, gamma=5):
    """Expected tokens per step given per-head acceptance rates."""
    expected = 0.0
    prob = 1.0
    for i, acc in enumerate(acceptance_rates[:gamma]):
        expected += prob * acc * 1.0
        prob *= acc
    return 1.0 + expected  # +1 for the verified target token

print("Medusa head acceptance rates (empirical from paper):")
accept_rates = [0.85, 0.72, 0.61, 0.52, 0.44]  # head 1-5
for i, ar in enumerate(accept_rates):
    print(f"  Head {i+1}: {ar:.0%} acceptance")

speedup = medusa_speedup(accept_rates)
print(f"\nExpected tokens per step: {speedup:.2f}")
print(f"Speedup over greedy: {speedup:.2f}x")
print(f"Tree candidates: {medusa_tree_candidates()} (pruned in practice to ~64)")

# Compare configurations
print("\nMedusa speedup vs number of heads:")
for k in [1, 2, 3, 4, 5]:
    sp = medusa_speedup(accept_rates, gamma=k)
    print(f"  K={k}: {sp:.2f}x speedup")
```
