## 1. Latency Decomposition

Profile each component of end-to-end latency to identify where time is spent.


```python
import numpy as np, matplotlib.pyplot as plt

def simulate_e2e_latency(prompt_len=512, output_len=256):
    components = {
        "Network (in)": np.random.normal(5, 1),
        "Tokenization": np.random.normal(1.5, 0.2),
        "Queue wait": np.random.exponential(8),
        "Prefill (TTFT)": prompt_len * np.random.normal(0.08, 0.01),
        "Decode (TBT)": output_len * np.random.normal(20, 2),
        "Detokenization": np.random.normal(0.5, 0.1),
        "Network (out)": np.random.normal(3, 1),
    }
    return components

samples = [simulate_e2e_latency() for _ in range(100)]
keys = list(samples[0].keys())
means = {k: np.mean([s[k] for s in samples]) for k in keys}
total = sum(means.values())
print("End-to-end latency breakdown (prompt=512, output=256):")
for k, v in means.items():
    pct = v/total*100
    print(f"  {k:<25} {v:>8.1f}ms {pct:>6.1f}%")
print(f"  {'Total':<25} {total:>8.1f}ms")
plt.figure(figsize=(10,5))
plt.pie(list(means.values()), labels=list(means.keys()), autopct='%1.0f%%')
plt.title('E2E Latency Breakdown'); plt.savefig('e2e_breakdown.png',dpi=100); plt.show()
```
