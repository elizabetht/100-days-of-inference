Compare a fine-tuned 7B model to a quantized 70B model for a specific task. The 70B INT4 has similar parameters-visible as a 17.5B FP16 model in terms of model capacity, but 10x the memory bandwidth pressure.

## 1. Fine-Tune vs Quantized Comparison

Model quality, latency, and cost tradeoffs for deployment of fine-tuned 7B vs quantized 70B.


```python
import numpy as np

models = [
    # name, params_b, quantization, task_quality, latency_ms, gpu_mem_gb, cost_per_hr
    ("7B FT (LoRA, FP16)", 7, "FP16", 0.91, 22, 14, 0.5),
    ("7B INT4", 7, "INT4", 0.87, 12, 4, 0.5),
    ("13B FP16", 13, "FP16", 0.93, 38, 26, 0.5),
    ("70B INT4", 70, "INT4", 0.97, 95, 35, 2.0),
    ("70B FP16", 70, "FP16", 0.99, 310, 140, 8.0),
]

print(f"{'Model':<22} {'Quality':>9} {'Latency':>9} {'VRAM':>7} {'$/hr':>6} {'$/1M tok':>10}")
for name, params, quant, quality, lat_ms, vram, cost in models:
    tps = 1000 / lat_ms  # tokens/sec at bs=1
    cost_per_mtok = cost / (tps * 3600) * 1e6
    print(f"{name:<22} {quality:>9.2f} {lat_ms:>8}ms {vram:>6}GB {cost:>6.1f} {cost_per_mtok:>10.2f}")
print()
print("Recommendation:")
print("  Task-specific 7B FT: lowest latency + cost, ~92% of 70B quality")
print("  70B INT4: best quality for general tasks, 4x cost")
print("  Fine-tune when task domain is narrow and latency matters.")
```
