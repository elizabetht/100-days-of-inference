## 1. TP Scaling Analysis

Ideal linear scaling is limited by AllReduce communication. Measure where communication overhead dominates.


```python
import numpy as np, matplotlib.pyplot as plt

def tp_scaling(tp_sizes, compute_ms=100, allreduce_ms_per_level=5):
    """Model TP throughput scaling."""
    results = []
    for n in tp_sizes:
        compute = compute_ms / n
        comm = allreduce_ms_per_level * np.log2(n) if n > 1 else 0
        total = compute + comm
        speedup = compute_ms / total
        results.append({"n":n,"total_ms":total,"speedup":speedup,"efficiency":speedup/n})
    return results

results = tp_scaling([1,2,4,8,16])
print(f"{'TP':>5} {'Latency (ms)':>15} {'Speedup':>10} {'Efficiency':>12}")
for r in results:
    print(f"{r['n']:>5} {r['total_ms']:>15.1f} {r['speedup']:>10.2f}x {r['efficiency']:>11.0%}")
np.random.seed(42)
throughput = [r['speedup']*100 + np.random.normal(0,5) for r in results]
plt.plot([r['n'] for r in results], throughput, 'b-o')
plt.plot([r['n'] for r in results], [r['n']*100 for r in results], 'r--', label='Ideal')
plt.xlabel("TP Degree"); plt.ylabel("Throughput (tokens/s)")
plt.title("TP Scaling: Actual vs Ideal"); plt.legend(); plt.grid(True)
plt.savefig("tp_scaling_empirical.png",dpi=100); plt.show()
```
