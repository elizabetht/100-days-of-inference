## Concept Overview

Text-to-speech inference for real-time applications requires streaming: generating and playing audio chunks before the full text is synthesized. The key metric is time-to-first-audio (TTFA).


```python
import torch
import torch.nn as nn
import torch.nn.functional as F
import numpy as np
import matplotlib.pyplot as plt
import time

print(f'CUDA: {torch.cuda.is_available()}')
if torch.cuda.is_available(): print(f'GPU: {torch.cuda.get_device_name(0)}')
```

## 1. TTS Latency Model

TTS latency has two components: text encoding (fast) and vocoder inference (slow). Streaming sends audio chunks as they're generated.


```python
import numpy as np

def tts_streaming_latency(text_chars, chars_per_phoneme=4, phonemes_per_chunk=20,
                          phoneme_ms=5, vocoder_ms_per_chunk=50):
    n_phonemes = text_chars // chars_per_phoneme
    n_chunks = int(np.ceil(n_phonemes / phonemes_per_chunk))
    ttfa = phoneme_ms * phonemes_per_chunk + vocoder_ms_per_chunk
    total = n_chunks * (phoneme_ms * phonemes_per_chunk + vocoder_ms_per_chunk)
    return ttfa, total

print("TTS streaming: time to first audio vs text length")
print(f"{'Text (chars)':>14} {'TTFA (ms)':>12} {'Total (ms)':>12}")
for chars in [50, 100, 200, 500, 1000]:
    ttfa, total = tts_streaming_latency(chars)
    print(f"{chars:>14} {ttfa:>12.0f} {total:>12.0f}")
print()
print("Best practice: stream TTS output while LLM is still generating text.")
print("Speech-to-speech pipeline: ASR -> LLM -> TTS each stage can be streamed.")
```
