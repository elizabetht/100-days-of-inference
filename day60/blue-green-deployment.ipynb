## 1. Traffic Switch

Blue-green swaps traffic atomically after green passes health validation.


```python
class BlueGreen:
    def __init__(self):
        self.state = "blue"
        self.blue_traffic = 1.0

    def deploy_green(self): print("Deploying green (v2)...")
    def health_check(self, env):
        checks = [("GPU memory OK", True), ("Error rate < 0.1%", True), ("P99 TTFT < 200ms", True)]
        all_ok = all(v for _,v in checks)
        print(f"{env} health checks: {'ALL PASS' if all_ok else 'FAIL'}")
        return all_ok
    def switch(self):
        self.state = "green"
        print("Traffic switched to green (v2)")
        print("Blue (v1) standing by for 10min, then teardown")

bg = BlueGreen()
bg.deploy_green()
if bg.health_check("green"):
    bg.switch()
    print(f"State: {bg.state}")
```
