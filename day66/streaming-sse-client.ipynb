## 1. SSE Streaming Client

A streaming client using httpx reads token deltas and measures TTFT from first chunk arrival.


```python
import time, json

def simulate_sse_response():
    """Simulate SSE chunks from server."""
    import time
    tokens = ["This", " is", " a", " streaming", " response"]
    for i, tok in enumerate(tokens):
        time.sleep(0.02 if i == 0 else 0.01)
        yield json.dumps({"choices":[{"delta":{"content":tok},"finish_reason":None}]})
    yield "[DONE]"

def streaming_client_with_metrics(prompt):
    ttft = None
    tokens = []
    start = time.perf_counter()
    for chunk in simulate_sse_response():
        now = time.perf_counter()
        if chunk == "[DONE]": break
        data = json.loads(chunk)
        content = data["choices"][0]["delta"].get("content", "")
        if content:
            if ttft is None:
                ttft = (now - start) * 1000
            tokens.append(content)
    return {"ttft_ms": ttft, "tokens": tokens, "count": len(tokens)}

result = streaming_client_with_metrics("Hello")
print(f"TTFT: {result['ttft_ms']:.1f} ms")
print(f"Tokens: {result['tokens']}")
print(f"Token count: {result['count']}")
```
