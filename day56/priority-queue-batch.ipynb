```python
import heapq, time, numpy as np

class PriorityBatchQueue:
    def __init__(self, max_batch=8):
        self.queues = {0: [], 1: []}  # 0=high, 1=low priority
        self.max_batch = max_batch
        self.counter = 0

    def enqueue(self, request, priority=1):
        heapq.heappush(self.queues[priority], (time.time(), self.counter, request))
        self.counter += 1

    def dequeue_batch(self):
        batch = []
        # Drain high priority first
        while self.queues[0] and len(batch) < self.max_batch:
            _, _, req = heapq.heappop(self.queues[0])
            batch.append(('high', req))
        # Fill remaining slots with low priority
        while self.queues[1] and len(batch) < self.max_batch:
            _, _, req = heapq.heappop(self.queues[1])
            batch.append(('low', req))
        return batch

pq = PriorityBatchQueue(max_batch=8)
import random; random.seed(42)
for i in range(20):
    pq.enqueue({'id': i, 'tokens': random.randint(10,200)}, priority=random.choice([0,0,0,1]))

print("Priority queue batch formation:")
for _ in range(3):
    batch = pq.dequeue_batch()
    if batch:
        hp = [r['id'] for p,r in batch if p=='high']
        lp = [r['id'] for p,r in batch if p=='low']
        print(f"  Batch: {len(batch)} reqs | high={hp} low={lp}")
```
