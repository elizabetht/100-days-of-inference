## 1. Geo-Aware Routing

Assign each request to the pool with lowest latency that meets cost and availability constraints.


```python
import numpy as np

pools = [
    {"name":"spark-local","lat":0.5,"avail":0.99,"cost":0},
    {"name":"aws-us-east","lat":25,"avail":0.99,"cost":10},
    {"name":"gcp-us-central","lat":30,"avail":0.99,"cost":9.5},
    {"name":"aws-eu-west","lat":90,"avail":0.99,"cost":10.5},
]

def geo_route(user_lat_ms, latency_slo_ms=100, max_cost=12):
    eligible = [p for p in pools if p["avail"]>0.95 and
                abs(p["lat"]-user_lat_ms)<latency_slo_ms and p["cost"]<=max_cost]
    if not eligible: eligible = pools
    return min(eligible, key=lambda p: abs(p["lat"]-user_lat_ms))

print("Geo-aware routing decisions:")
for user_lat, region in [(2,"local user"),(40,"East Coast"),(100,"Europe"),(200,"Asia")]:
    pool = geo_route(user_lat)
    print(f"  {region:<15} (lat={user_lat}ms) → {pool['name']} (est_lat={abs(pool['lat']-user_lat):.0f}ms)")
```
