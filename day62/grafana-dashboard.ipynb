## 1. Dashboard Design

Standard inference Grafana dashboard with alert rules.


```python
import json
dashboard = {
    "title": "LLM Inference Monitoring",
    "panels": [
        {"title":"TTFT P99","type":"timeseries","targets":[{"expr":"histogram_quantile(0.99, inference_ttft_seconds_bucket)"}],"alert":{"threshold":0.5,"condition":">"}},
        {"title":"Token Throughput","type":"stat","targets":[{"expr":"rate(inference_tokens_total[1m])"}]},
        {"title":"KV Cache Utilization","type":"gauge","targets":[{"expr":"inference_kv_cache_utilization"}],"thresholds":[0.7,0.9]},
        {"title":"Queue Depth","type":"timeseries","targets":[{"expr":"inference_queue_depth"}]},
        {"title":"GPU Utilization","type":"heatmap","targets":[{"expr":"inference_gpu_utilization"}]},
        {"title":"Error Rate","type":"timeseries","targets":[{"expr":"rate(inference_errors_total[5m])"}]},
    ]
}
print("Grafana dashboard JSON (summary):")
for panel in dashboard["panels"]:
    alert = panel.get("alert",{})
    print(f'  {panel["title"]:<30} {panel["type"]:<15} alert: {alert.get("condition","")}{alert.get("threshold","")}')  
```
