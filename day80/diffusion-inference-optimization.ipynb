## Concept Overview

Diffusion models run N denoising steps (typically 20-50 DDIM steps). Each step is a full UNet forward pass. Optimizations: fewer steps (SDXL-Turbo: 4 steps), DPM++ scheduler, INT8 quantization, Flash Attention in attention layers.

## 1. Diffusion Step Budget

Total latency = steps * unet_time. SDXL UNet ~250ms/step on A100. Target: <2s for 8 steps.


```python
import numpy as np, matplotlib.pyplot as plt

def diffusion_latency(steps, unet_ms=250, scheduler="DDIM"):
    """Estimate diffusion model latency."""
    # DPM++ 2M needs 2 UNet evals per step after step 1
    if scheduler == "DPM++2M":
        total_evals = 1 + 2*(steps-1)
    else:  # DDIM
        total_evals = steps
    return total_evals * unet_ms

print("Diffusion latency vs step count:")
print(f"{'Steps':>8} {'DDIM (ms)':>12} {'DPM++ 2M (ms)':>16}")
for s in [4, 8, 12, 20, 30, 50]:
    d = diffusion_latency(s, "DDIM")
    p = diffusion_latency(s, "DPM++2M")
    print(f"{s:>8} {d:>12.0f} {p:>16.0f}")
print()
print("Optimization stack: SDXL-Turbo (4 steps) + TRT + Flash Attn + INT8 UNet")
print("Target: <500ms for 512x512 @ 4 steps on H100")
```
