## Concept Overview

RoPE (Rotary Position Embedding) encodes relative positions via rotation matrices. Scaling the RoPE base frequency (base=10000->500000+) allows models to generalize to contexts longer than training.


```python
import torch
import torch.nn as nn
import torch.nn.functional as F
import numpy as np
import matplotlib.pyplot as plt
import time

print(f'CUDA: {torch.cuda.is_available()}')
if torch.cuda.is_available(): print(f'GPU: {torch.cuda.get_device_name(0)}')
```

## 1. RoPE Position Encoding

Compute RoPE frequencies and show how scaling extends to longer contexts.


```python
import numpy as np, matplotlib.pyplot as plt

def rope_freqs(d_model, base=10000, max_seq=4096):
    theta = 1.0 / (base ** (np.arange(0, d_model, 2) / d_model))
    positions = np.arange(max_seq)
    freqs = np.outer(positions, theta)
    return np.cos(freqs), np.sin(freqs)

fig, axes = plt.subplots(1, 3, figsize=(15,4))
bases = [10000, 100000, 500000]
titles = ["Base 10K (Llama-2)", "Base 100K (Llama-3)", "Base 500K (Qwen2)"]
for ax, base, title in zip(axes, bases, titles):
    cos_f, _ = rope_freqs(128, base=base, max_seq=8192)
    ax.imshow(cos_f.T, aspect="auto", cmap="RdBu")
    ax.set_title(title); ax.set_xlabel("Position"); ax.set_ylabel("Dim pair")
plt.tight_layout()
plt.savefig("rope_scaling.png", dpi=100); plt.show()

print("RoPE base frequency comparison:")
for base, ctx in [(10000,"4K"), (100000,"128K"), (500000,"2M")]:
    freqs = 1.0 / (base ** (np.arange(0, 128, 2) / 128))
    print(f"  base={base:>7,}: min_freq={freqs[-1]:.2e} -> max_ctx={ctx}")
```
