## 1. TP Scaling Model

Measure compute speedup vs communication overhead for different TP degrees.


```python
import numpy as np, matplotlib.pyplot as plt

def tp_efficiency(n, compute_s=1.0, allreduce_s_per_gpu=0.1):
    compute_speedup = n  # ideal linear
    allreduce_overhead = allreduce_s_per_gpu * (n-1)
    total = compute_s / n + allreduce_overhead
    efficiency = (compute_s / n) / total
    return efficiency, 1 / total

print("TP scaling efficiency:")
for n in [1,2,4,8]:
    eff, speedup = tp_efficiency(n)
    print(f"  TP={n}: efficiency={eff:.0%} speedup={speedup:.2f}x")

ns = range(1, 17)
effs = [tp_efficiency(n)[0] for n in ns]
plt.plot(ns, effs, 'b-o')
plt.xlabel("TP Degree"); plt.ylabel("Efficiency")
plt.title("Tensor Parallel Efficiency vs Degree"); plt.grid(True)
plt.savefig("tp_scaling.png", dpi=100); plt.show()
```
