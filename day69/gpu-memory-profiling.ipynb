## 1. Memory Budget Analysis

Compute memory budget for model weights, KV cache, and activations to find the maximum batch size.


```python
def memory_budget(model_params_b, gpu_gb, kv_heads=8, d_head=128, num_layers=32,
                  max_seq=8192, dtype_bytes=2):
    weights_gb = model_params_b * dtype_bytes
    kv_gb = 2 * num_layers * kv_heads * d_head * max_seq * dtype_bytes / 1e9
    activations_gb = 0.5  # rough estimate for intermediate activations
    available_for_batches = gpu_gb - weights_gb - kv_gb - activations_gb
    kv_per_seq_gb = 2 * num_layers * kv_heads * d_head * max_seq * dtype_bytes / 1e9
    max_batch = int(available_for_batches / kv_per_seq_gb) if kv_per_seq_gb > 0 else 0
    return {"weights_gb":weights_gb,"kv_gb":kv_gb,"available_gb":available_for_batches,"max_batch":max_batch}

print("Memory budget analysis (128GB GB10 Spark):")
for params, name in [(8,"Llama-3-8B"),(27,"Llama-3.1-27B")]:
    m = memory_budget(params, 128)
    print(f"{name}: weights={m['weights_gb']}GB kv={m['kv_gb']:.1f}GB avail={m['available_gb']:.1f}GB max_batch={m['max_batch']}")
```
