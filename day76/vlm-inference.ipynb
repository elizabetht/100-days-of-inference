## Concept Overview

VLM inference requires image preprocessing (resize, normalize, patch-encode) before the language model sees any tokens. The image encoder runs once per image; the LLM then treats image patch embeddings as prefix tokens.

## 1. Image Preprocessing Pipeline

Resize to 336x336, normalize with CLIP stats, and divide into 14x14 patches of 16px each (576 tokens per image).


```python
import numpy as np, matplotlib.pyplot as plt

def preprocess_image(H=336, W=336, patch_size=14):
    """Simulate VLM image preprocessing."""
    img = np.random.rand(3, H, W)
    # Normalize with CLIP mean/std
    mean = np.array([0.48145466, 0.4578275, 0.40821073])[:,None,None]
    std  = np.array([0.26862954, 0.26130258, 0.27577711])[:,None,None]
    img = (img - mean) / std
    # Patchify: (3, H, W) -> (n_patches, patch_size^2*3)
    n_h = H // patch_size
    n_w = W // patch_size
    n_patches = n_h * n_w
    patches = img.reshape(3, n_h, patch_size, n_w, patch_size)
    patches = patches.transpose(1,3,0,2,4).reshape(n_patches, -1)
    return patches

patches = preprocess_image()
print(f"Image 336x336 -> {patches.shape[0]} patches of dim {patches.shape[1]}")
print(f"Token cost: {patches.shape[0]} image tokens + system_prompt + user_prompt")
print()
# Compare models
for model, res, ptok in [("LLaVA-1.5","336x336",576), ("InternVL2","448x448",1024), ("Qwen-VL","448x448",256)]:
    print(f"  {model:<15} res={res} img_tokens={ptok}")
```
