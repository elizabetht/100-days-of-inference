## 1. Locust Load Test

Locust simulates concurrent users sending requests at a specified rate.


```python
print("Locust load test configuration:")
locustfile = """
from locust import HttpUser, task, between
import json

class InferenceUser(HttpUser):
    wait_time = between(0.1, 0.5)

    @task
    def chat_completion(self):
        payload = {
            "model": "llama-3-8b",
            "messages": [{"role": "user", "content": "Summarize AI in 3 sentences."}],
            "max_tokens": 100
        }
        with self.client.post("/v1/chat/completions", json=payload, catch_response=True) as r:
            if r.status_code == 200:
                r.success()
            else:
                r.failure(f"HTTP {r.status_code}")
"""
print(locustfile)
print("Run: locust -f locustfile.py --host http://192.168.1.76:8000")
print("Ramp: 1 user/s from 1 to 100")
print("Watch for: P99 TTFT >2x, error rate >1%  -- these mark saturation")
```
