## 1. OpenTelemetry Instrumentation

Each request gets a trace with child spans for tokenization, prefill, decode, and detokenization.


```python
try:
    from opentelemetry import trace
    from opentelemetry.sdk.trace import TracerProvider
    from opentelemetry.sdk.trace.export import SimpleSpanProcessor, ConsoleSpanExporter
    provider = TracerProvider()
    provider.add_span_processor(SimpleSpanProcessor(ConsoleSpanExporter()))
    trace.set_tracer_provider(provider)
    tracer = trace.get_tracer("inference-server")
    with tracer.start_as_current_span("inference_request") as root:
        with tracer.start_as_current_span("tokenization"): pass
        with tracer.start_as_current_span("prefill"): pass
        with tracer.start_as_current_span("decode"): pass
    print("OpenTelemetry trace exported")
except ImportError:
    print("pip install opentelemetry-sdk")
    print("Trace structure for inference request:")
    spans = [("inference_request",0,None),("tokenization",2,0),("prefill",8,0),
             ("decode",180,0),("detokenization",0.5,0)]
    for name,dur,parent in spans:
        indent = "  " if parent else ""
        print(f"{indent}{name}: {dur}ms")
```
