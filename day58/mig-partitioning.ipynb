## 1. MIG Configuration

Profile different MIG partition sizes and model them.


```python
print("MIG configuration commands:")
print("  sudo nvidia-smi mig -i 0 -cci")
print("  sudo nvidia-smi mig -cgi 1g.10gb,1g.10gb,2g.20gb,3g.40gb -C")
print("  nvidia-smi -L  # list instances")
print()
print("Instance profiles (H100 80GB):")
profiles = [('1g.10gb',7,10,475,124),('2g.20gb',4,20,950,247),('3g.40gb',2,40,1900,371),('7g.80gb',1,80,3350,989)]
for name,max_i,mem,bw,tflops in profiles:
    print(f"  {name:<10} max={max_i}x mem={mem}GB bw={bw}GB/s fp16={tflops}TF")
```
