## Concept Overview

Embedding models (BERT-family, E5, GTE) benefit enormously from batching because they run a single forward pass without autoregressive decoding. Throughput scales nearly linearly with batch size up to memory limits.


```python
import torch
import torch.nn as nn
import torch.nn.functional as F
import numpy as np
import matplotlib.pyplot as plt
import time

print(f'CUDA: {torch.cuda.is_available()}')
if torch.cuda.is_available(): print(f'GPU: {torch.cuda.get_device_name(0)}')
```

## 1. Embedding Throughput Benchmark

Unlike LLMs, embedding models don't have KV cache growth. Throughput is batch_size * seq_len / forward_pass_time.


```python
import time, numpy as np
try:
    import torch
    device = "cuda" if torch.cuda.is_available() else "cpu"
except ImportError:
    device = "cpu"

def simulate_embedding_throughput(batch_size, seq_len, d_model=768, n_layers=12):
    """Estimate embedding throughput (tokens/sec)."""
    # Attention is O(seq^2 * d_model) per layer
    flops = 2 * batch_size * seq_len**2 * d_model * n_layers
    gpu_tflops = 67e12  # DGX Spark ~67 TFLOPS FP16
    time_s = flops / gpu_tflops
    tokens = batch_size * seq_len
    return tokens / time_s

print(f"Embedding model throughput (BERT-base, seq=128):")
print(f"{'Batch Size':>12} {'Throughput (tok/s)':>20} {'Sentences/s':>14}")
for bs in [1, 8, 32, 128, 512, 2048]:
    tps = simulate_embedding_throughput(bs, 128)
    sps = tps / 128
    print(f"{bs:>12} {tps:>20,.0f} {sps:>14,.0f}")
print()
print("Key insight: embedding models are pure throughput — maximize batch size.")
```
