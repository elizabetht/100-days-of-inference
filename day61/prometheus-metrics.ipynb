## 1. Metrics Server

Expose TTFT, TPOT, token count, GPU util, and error rate as Prometheus metrics.


```python
try:
    from prometheus_client import Counter, Histogram, Gauge, start_http_server
    req_counter = Counter("inference_requests_total", "Total inference requests")
    ttft_hist = Histogram("ttft_seconds", "Time to first token", buckets=[0.05,0.1,0.2,0.5,1.0,2.0])
    gpu_util = Gauge("gpu_utilization", "GPU utilization 0-1")
    print("Prometheus metrics registered:")
    for m in [req_counter, ttft_hist, gpu_util]:
        print(f"  {m._name}")
    # start_http_server(9090)  # Uncomment to expose /metrics
    print("To expose: start_http_server(9090)")
except ImportError:
    print("pip install prometheus_client")
    print("Metrics to expose:")
    metrics = [("inference_requests_total","counter","Requests served"),
               ("ttft_seconds","histogram","TTFT distribution"),
               ("tpot_seconds","histogram","TPOT distribution"),
               ("kv_cache_utilization","gauge","KV cache fill %"),
               ("queue_depth","gauge","Pending requests")]
    for name, mtype, desc in metrics:
        print(f"  {name:<35} {mtype:<10} {desc}")
```
